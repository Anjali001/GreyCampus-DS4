```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shopping_cart = { 'fruits':500 ,'Snacks':1000, 'Grocery':500,'Chocolates':200,'Toffee':30}
item = list(shopping_cart.keys())
worth = list(shopping_cart.values())
```

```python
print(item)
print(worth)
```

```python
plt.figure(figsize=(10,5))
plt.bar(item,worth,color='grey')
plt.yticks(np.arange(0,1001,100))
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x=item,y=worth,palette='viridis')
plt.show()
```
