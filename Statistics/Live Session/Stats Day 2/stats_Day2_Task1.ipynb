```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
'Country' : ['India','India','US','US','US','US','Canada','Canada','Russia','UK'],
'Name' : ['A','B','C','D','E','F','G','H','I','J'],
'Age' : [32,10,30,45,67,21,31,56,78,20],
'Weight' : [60,70,70,54,89,100,105,48,50,67],
'Sex' : ['M','M','F','F','F','M','F','F','M','M']
})
```

```python
sns.boxplot(data = data , x= 'Age')
```

There are no outliers. It seems like the data is right or positively skewed. Median is around 31.

```python
sns.scatterplot(data = data , x= 'Age', y='Weight',hue='Sex')
```

No pattern is clearly visible. We plot scatterplot between two quantitative variables and can add a categorical variable to add more details to this.

```python
sns.histplot(data = data, x='Age')
```

As we can see there are bins like 10-24, 24-37,37-51 etc. This shows that there are 3 people between 10-24, 3 people between 24-37 and so on.

So, histogram shows the frequency distribution of a variable.

```python
y = data['Country'].value_counts()
```

```python
plt.pie(y,autopct="%.1f%%",labels=y.index)
plt.show()
```

Pie chart shows the percentage distribution of variable. Eg 40 % of the data is from US.

```python
sns.barplot(data=data,x='Country',y='Age')
```

This barplot shows the average Age for a specific country. The line is middle tells about confidence limits.
